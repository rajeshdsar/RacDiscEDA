### Question 1: What test is appropriate for this problem? Does CLT apply?

The question here is find out if there is a difference in the average * call rate * between White race applicants and other race applicants. So, we are comparing the mean of two samples, $X bar$, where the population standard deviation &sigma; is not known. The sample size, $n$ is > 30. Considering all these, we will use a two sample $t$ test to solve this. Yes, Central Limit Theorem does apply.

### Question 3: Compute margin of error, confidence interval, and p-value

```python
data_b=data[data.race=='b']
```

```python
data_b.head(5)
```

```python
len(data_b)
```

```python
data_w=data[data.race=='w']
```

```python
data_w.head(5)
```

```python
len(data_w)
```

```python
b_calls = data_b.call
```

```python
w_calls = data_w.call
```

```python
#mean of both samples and their difference - (0.09650924056768417, 0.0644763857126236, 0.03203285485506058)
w_calls.mean(),b_calls.mean(),w_calls.mean()-b_calls.mean()
```

```python
#standard deviation of both samples - (0.2953455150127411, 0.24564945697784424)
w_calls.std(),b_calls.std()
```

_ margin of error =  $t$<sub>critical</sub> $X$ $Standard Error$ _

_Standard Error = &radic;  ( (variance<sub>w</sub> / n<sub>w</sub>) + (variance<sub>b</sub> / n<sub>b</sub> ] _

The Degrees of Freedom for this is 4711

The $t$<sub>critical</sub> value is 1.960

```python
standard_error = math.sqrt(w_calls.var()/len(w_calls) + b_calls.var()/len(b_calls))
```

```python
standard_error * 1.960
```

confidence_interval = +- 0.01526, i.e. any difference above or below this interval will lead to the rejection of $H$<sub>o</sub>. In this case the difference in mean between the two samples is 0.032, so $H$<sub>o</sub> can be rejected

```python
0.03203285485506058/ standard_error
```

```python
stats.ttest_ind(w_calls,b_calls)
```

** The $p$ value is 3.9408021031288859e-05 **
